```python
def remove_duplicates(DF):
    if DF.duplicated().any() == True :
        print("There are ",DF.duplicated().sum()," in the dataset!")
        DF.drop_duplicates()
        print("Duplicates Removed.")
    else:
        print("No duplicates in the dataset!")
```

```python
def is_float(x):
    try :
        float(x)
    except :
        return False
    return True   
```

```python
Banglore_house_DF_new_1[~Banglore_house_DF_new_1['total_sqft'].apply(is_float)].head(50)
```

```python
def convert_sqft_to_num(x):
    token = x.split('-')
    if len(token) == 2:
        return (float(token[0])+float(token[1]))/2
    try: 
        return float(x)
    except ValueError:
        if len(re.findall(r'(\d*\D*\d+)(\D+)', x)[0]) == 2:
            res = re.findall(r'(\d*\D*\d+)(\D+)', x)[0]
            if res[1] == 'Sq. Meter': 
                return round(float(res[0])*10.7639,3)
            elif res[1] == 'Sq. Yards':
                return round(float(res[0])*9,3)
            elif res[1] == 'Perch':
                return round(float(res[0])*272.25,3)
            elif res[1] == 'Acres':
                return round(float(res[0])*43560,3)
            elif res[1] == 'Cents':
                return round(float(res[0])*435.56,3)
            elif res[1] == 'Guntha':
                return round(float(res[0])*1089, 3)
            elif res[1] == 'Grounds':
                return round(float(res[0])*2400.35203, 3)
```

```python
Banglore_house_DF_new_2 = Banglore_house_DF_new_1.copy()
Banglore_house_DF_new_2['total_sqft'] = Banglore_house_DF_new_2['total_sqft'].apply(convert_sqft_to_num)
Banglore_house_DF_new_2.head(10)
```
