```python
def remove_duplicates(DF):
    if DF.duplicated().any() == True :
        print("There are ",DF.duplicated().sum()," in the dataset!")
        DF.drop_duplicates()
        print("Duplicates Removed.")
    else:
        print("No duplicates in the dataset!")
```

```python
remove_duplicates(Banglore_house_DF_new_1)
```

```python
def uni_val(DF):
    for i in DF.columns:
        if DF[i].dtype == 'O':
            print("\n")
            print(f"Unique values of {i} :\n")
            print(DF[i].unique())
            print("\n")
```
